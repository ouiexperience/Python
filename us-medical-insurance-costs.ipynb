```python
list_of_ages =[]
list_of_gender =[]
list_of_bmi =[]
list_of_children =[]
list_of_smoker =[]
list_of_region =[]
list_of_charges =[]
with open(r"C:\Users\anian\OneDrive\Desktop\Ani\DATA Analytics\Codeacademy\Python\US medical insurance cost\python-portfolio-project-starter-files\python-portfolio-project-starter-files\insurance.csv") as users_medical:
    user_reader = csv.DictReader(users_medical)
    for i in user_reader:
        list_of_ages.append(i["age"])  
        list_of_gender.append(i["sex"])
        list_of_bmi.append(i["bmi"])
        list_of_children.append(i["children"])
        int_list_of_children = [int(x) for x in list_of_children]
        list_of_smoker.append(i["smoker"])
        list_of_region.append(i["region"])
        list_of_charges.append(i["charges"])
        
```

```python
#Average number of children per patient in insurance.csv
total_adults = len(list_of_children)
total_children = sum(int_list_of_children)
children_per_adult =total_children/total_adults
print("The average number of children per adult is " + str(children_per_adult))
```

```python
# Percentage of patients in insurance.csv in each region 

def region(list_of_region):
    southwest = 0
    northwest = 0
    southeast = 0
    northeast = 0

    for i in list_of_region:
        if i =="southwest":
            southwest +=1
        elif i == "northwest":
            northwest+= 1
        elif i == "southeast":
            southeast +=1
        elif i == "northeast":
            northeast +=1
            
    return southwest, northwest, southeast, northeast


southwest, northwest, southeast, northeast = region(list_of_region)

print (str(southwest/total_adults*100) + " % of people live in the Southwest region")
print (str(northwest/total_adults*100) + " % of people live in the Northwest region")
print (str(southeast/total_adults*100) + " % of people live in the Southeast region")
print (str(northeast/total_adults*100) + " % of people live in the Northeast region")
```

```python
#As observed the percentage of people living in the 4 regions are pretty close to each other (The southeast has a little more but I think it is negligible)
```

```python
#what is the percentage of smokers verses non smokers for our patients in insurance.csv
def smoker(list_of_smoker):
    smoker = 0
    non_smoker = 0
    for i in list_of_smoker:
        if i == "yes":
            smoker +=1
        elif i == "no":
            non_smoker +=1
    return smoker, non_smoker 


smoker, non_smoker = smoker(list_of_smoker)


print(str(smoker/total_adults * 100) + " % of people are smokers")
print(str(non_smoker/total_adults * 100) + " % of people are non smokers")
print(smoker)
print(non_smoker)
```

```python
# A small portion of the population smokes 20.5% compared to 79.5% of non smoking people
```

```python
#what percentage of the patients in insurance.csv that are male and what percentage are female?
def gender(list_of_gender):
    female = 0
    male = 0
    for i in list_of_gender:
        if i == "female":
            female +=1
        elif i == "male":
            male +=1
    return female, male 


female, male = gender(list_of_gender)


print(str(female/total_adults * 100) + " % of people are female")
print(str(male/total_adults * 100) + " % of people are male")
```

```python
#female and male percentage are also almost equal.
```

```python
#How old are the patients in insurance.csv
list_of_ages = [float(ages) for ages in list_of_ages]
def age(list_of_ages):
    under_18 = 0
    Bet18_19 = 0
    Bet20_24 = 0
    Bet25_29 = 0
    Bet30_34 =0
    Bet35_39 = 0
    Bet40_44 = 0
    Bet45_49 = 0
    Bet50_54 = 0
    Bet55_59 = 0
    over_60 = 0
    for i in list_of_ages:
        if i < 18:
            under_18 +=1
        elif i >=18 and i <20:
            Bet18_19 +=1
        elif i >=20 and i <25:
            Bet20_24 +=1
        elif i >=25 and i <30:
            Bet25_29 +=1
        elif i >=30 and i <35:
            Bet30_34 +=1
        elif i >=35 and i <40:
            Bet35_39 +=1
        elif i >=40 and i <45:
            Bet40_44 +=1
        elif i >=45 and i <50:
            Bet45_49 +=1
        elif i >=50 and i <55:
            Bet50_54 +=1
        elif i >=55 and i <60:
            Bet55_59 +=1
        elif i >=60:
            over_60 +=1
            
            
    return under_18, Bet18_19, Bet20_24, Bet25_29, Bet30_34, Bet35_39, Bet40_44, Bet45_49, Bet50_54, Bet55_59, over_60 

under_18, Bet18_19, Bet20_24, Bet25_29, Bet30_34, Bet35_39, Bet40_44, Bet45_49, Bet50_54, Bet55_59, over_60  = age(list_of_ages)

print(str(under_18/total_adults * 100) + " % of people are under 18 years old")
print(str(Bet18_19/total_adults * 100) + " % of people are between 18 and 19 years old")
print(str(Bet20_24/total_adults * 100) + " % of people are between 20 and 24 years old")
print(str(Bet25_29/total_adults * 100) + " % of people are between 25 and 29 years old")   
print(str(Bet30_34/total_adults * 100) + " % of people are between 30 and 34 years old")    
print(str(Bet35_39/total_adults * 100) + " % of people are between 35 and 39 years old")
print(str(Bet40_44/total_adults * 100) + " % of people are between 40 and 44 years old")   
print(str(Bet45_49/total_adults * 100) + " % of people are between 45 and 49 years old")    
print(str(Bet50_54/total_adults * 100) + " % of people are between 50 and 54 years old")
print(str(Bet55_59/total_adults * 100) + " % of people are between 55 and 59 years old")    
print(str(over_60/total_adults * 100) + " % of people are over 60 years old")
```

```python
# calculate the averge BMI's
list_of_bmi = [float(bmi) for bmi in list_of_bmi]
def BMI(list_of_bmi):
    low_bmi = 0
    normal_bmi = 0
    overweight_bmi = 0
    obese_bmi = 0
    severe_bmi = 0
    top_bmi = 0
    for i in list_of_bmi:
        if i < 18.5:
             low_bmi += 1
        elif i >= 18.5 and i <= 24.9:
             normal_bmi += 1   
        elif i > 24.9 and i <= 29.9:
            overweight_bmi += 1    
        elif i > 29.9 and i <= 34.9:
            obese_bmi += 1    
        elif i > 34.9 and i <= 39.9:
            severe_bmi += 1
        elif i > 39.9:
            top_bmi += 1
    return low_bmi, normal_bmi, overweight_bmi, obese_bmi, severe_bmi, top_bmi

low_bmi, normal_bmi, overweight_bmi, obese_bmi, severe_bmi, top_bmi = BMI(list_of_bmi)

print(str(low_bmi/total_adults * 100) + " % of people the have a BMI under 18.5")
print(str(normal_bmi/total_adults * 100) + " % of people that have a BMI between 18.5 and 24.9")
print(str(overweight_bmi/total_adults * 100) + " % of people that have a BMI between 24.9 and 29.9")
print(str(obese_bmi/total_adults * 100) + " % of people that have a BMI between 29.9 and 34.9")   
print(str(severe_bmi/total_adults * 100) + " % of people that have a BMI between 34.9 and 39.9")    
print(str(top_bmi/total_adults * 100) + " % of people the have a BMI over 39.9")

```

# How does smoking affect the medical charges

```python
# I am assuming that smokers have bigger charges than non smokers
#I want to calculate the agerage charge or a smoker versus the average charge of a non smoker
smoker_charges =[]
non_smoker_charges =[]
for i in range(len(list_of_charges)):
    if list_of_smoker[i] == "yes":
        smoker_charges.append(float(list_of_charges[i]))
    elif list_of_smoker[i] == "no":
        non_smoker_charges.append(float(list_of_charges[i]))

average_smoker_charges = sum(smoker_charges) / len(smoker_charges) 
average_non_smoker_charges = sum(non_smoker_charges)/len(non_smoker_charges)
print ("The average charge for a smoker is $" +str(average_smoker_charges))
print ("The average charge for a non smoker is $" +str (average_non_smoker_charges))


```

```python
# The difference in charge for a smoker versus a non smoker is very considerable. Smoker pay in average 280% more than non smokers
```
